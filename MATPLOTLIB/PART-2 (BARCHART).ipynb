```python
language_counter=Counter()
for i in data["LanguagesWorkedWith"]:
    list=i.split(';')
    language_counter.update(list) 

languages=[]
popularity=[]
for i in language_counter.most_common(15):
    languages.append(i[0])
    popularity.append(i[1])
    
```

```python
print(languages)
print(popularity)
```

```python
plt.bar(languages,popularity)
plt.show()
```
