```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use("fivethirtyeight")
```

```python
data=pd.read_csv(r"C:\Users\Ranjithbalan\Downloads\data (2).csv")
data.head()
```

```python
age=data['Age']
all_dev=data["All_Devs"]
py=data['Python']
js=data["JavaScript"]

con_val=pd.concat([all_dev,py,js])
con_median=con_val.median()
```

```python
con_median
```

# line plot

```python
plt.plot(age,all_dev,label="All_dev")
plt.plot(age,py,label="Python")
plt.plot(age,js,label="JavaScript")
plt.show()
```

```python
plt.fill_between(age,py,con_median)
plt.show()
```

```python
plt.fill_between(age,all_dev,con_median,alpha=0.25)
plt.show()
```

```python
plt.fill_between(age,py,con_median,
                 where=(py> con_median),interpolate=True)
plt.fill_between(age,py,con_median,
                 where=(py<= con_median),interpolate=True)
plt.show()
```

```python
plt.fill_between(age,py,con_median,alpha=0.25,
                 where=(py> con_median),interpolate=True)
plt.fill_between(age,py,con_median,alpha=0.25,
                 where=(py<= con_median),interpolate=True)
plt.show()
```

```python
plt.fill_between(age,py,con_median,alpha=0.25,color="r",
                 where=(py> con_median),interpolate=True)
plt.fill_between(age,py,con_median,alpha=0.25,color='b',
                 where=(py<= con_median),interpolate=True)

plt.show()
```

```python
plt.fill_between(age,py,con_median,alpha=0.25,color="r",
                 where=(py> con_median),interpolate=True,label="above avg")
plt.fill_between(age,py,con_median,alpha=0.25,color='b',
                 where=(py<= con_median),interpolate=True,label="below avg")
plt.legend(loc=(0.0500000,0.8120000))
plt.show()
```

```python
plt.fill_between(age,py,all_dev,alpha=0.25,color="r",
                 where=(py> con_median),interpolate=True,label="above avg")
plt.fill_between(age,py,all_dev,alpha=0.25,color='b',
                 where=(py<= con_median),interpolate=True,label="below avg")
plt.legend(loc=(0.0500000,0.8120000))
plt.show()
```
