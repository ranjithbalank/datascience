```python
from matplotlib import pyplot as plt
from datetime import datetime,timedelta

from matplotlib import dates as mpl_dates
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot_date(dates,y)
plt.tight_layout()
plt.show()
```

```python
plt.plot_date(dates,y,linestyle='solid')
plt.tight_layout()
plt.show()
```

```python
plt.plot_date(dates,y,linestyle="solid",marker="None")
plt.tight_layout()
plt.show()

```

```python
plt.plot_date(dates,y,linestyle="--",marker="None")
plt.gcf().autofmt_xdate()
plt.show()
```
